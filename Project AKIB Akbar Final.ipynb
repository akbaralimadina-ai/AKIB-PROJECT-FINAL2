```python
%%writefile app.py
import streamlit as st
import google.generativeai as genai

# ================================
# KONFIGURASI MODEL & API
# ================================
genai.configure(api_key="AIzaSyBkOg4BlKimw_Rrj9o7jd219qINf4fm-BY")
model = genai.GenerativeModel("gemini-2.0-flash")

# ================================
# KONFIGURASI UI
# ================================
st.set_page_config(page_title="AKIB Chatbot", page_icon="ğŸŸ¢")
st.title("ğŸŸ¢ AKIB â€” Asisten Karya Ilmiah Birokrasi")
st.caption("Chatbot AI untuk ASN menyusun karya tulis ilmiah")

# ================================
# SESSION STATE
# ================================
if "messages" not in st.session_state:
    st.session_state.messages = []

# Tampilkan riwayat chat
for m in st.session_state.messages:
    st.chat_message(m["role"]).write(m["content"])

# ================================
# INPUT USER
# ================================
user_input = st.chat_input("Ketik pesan Anda...")

if user_input:
    st.session_state.messages.append({"role": "user", "content": user_input})
    st.chat_message("user").write(user_input)

    # PROMPT PREMIUM
    prompt = f"""
    Kamu adalah AKIB â€” Asisten Karya Ilmiah Birokrasi.

TUGAS UTAMA:
1. Membantu ASN menyiapkan karya tulis ilmiah:
   - Makalah
   - Artikel
   - Policy brief
   - Laporan analisis kebijakan
   - Jurnal ilmiah

2. Jika user baru mulai:
   - Sapa dengan ramah
   - Minta: Nama, Jabatan, Tupoksi

3. Setelah data diterima:
   - Berikan 5 rekomendasi judul ilmiah yang relevan
   - Minta memilih salah satu

4. Jika user memilih judul:
   - Tawarkan: Outline, Bab Iâ€“V, Referensi, atau Draft lengkap

5. Jika user meminta Bab Iâ€“V:
   - Buat lengkap, rapi, formal, bahasa ilmiah

6. Referensi:
   - Sertakan minimal 3 sumber ilmiah
   - Format APA Style

    User berkata: {user_input}
    """

    response = model.generate_content(prompt).text

    st.session_state.messages.append({"role": "assistant", "content": response})
    st.chat_message("assistant").write(response)
```
